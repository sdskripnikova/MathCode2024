```python
#задание 3
#функция 1 - airy - Воздушные функции и их производные 

import numpy as np
import matplotlib.pyplot as plt
from scipy.special import airy


x0 = 5
x1 = 15
x = np.linspace(x0, x1, 200)

# значения функции Эйри
ai, aip, bi, bip = airy(x) 

plt.plot(x, ai, label='Ai(x)')
plt.plot(x, bi, label='Bi(x)')
plt.plot(x, aip, label="Ai'(x)")
plt.plot(x, bip, label="Bi'(x)")
plt.xlabel('x')
plt.ylabel('y - значения функции ')
plt.title('airy - Воздушные функции и их производные. ')
plt.legend()
plt.show()
```

```python
#задание 3
#функция 2 - jvp(v, z[, n]) - Вычисление производных функций Бесселя первого рода.  

import numpy as np
import matplotlib.pyplot as plt
from scipy.special import jvp

v = np.array([0, 1, 2])
z = np.linspace(0, 10, 200)


jvp1 = jvp(v[0], z)
jvp2 = jvp(v[1], z)
jvp3 = jvp(v[2], z)


plt.plot(z, jvp1, label="при v=0")
plt.plot(z, jvp2, label="при v=1")
plt.plot(z, jvp3, label="при v=2")
plt.xlabel("z")
plt.ylabel("J'")
plt.title('jvp - Вычисление производных функций Бесселя первого рода. ')
plt.legend()
plt.show()
```

```python
import numpy as np

#Функция для интеграла
def f(y):
    return 1/(y**2-2*y-8)

#Метод прямоугольников
def rectangle_method(f, a, b, N):
    delta_x = (b - a) / N
    x_k = np.linspace(a, b, N+1)
    integral = delta_x * np.sum(f(x_k[:-1]))
    return integral

#Метод трапеций 
def trapezoid_method(f, a, b, N):
    delta_x = (b - a) / N
    x_k = np.linspace(a, b, N+1)
    integral = (delta_x / 2) * (f(a) + 2*np.sum(f(x_k[1:-1])) + f(b))
    return integral

#Правило Симпсона
def simpson_method(f, a, b, N):
    delta_x = (b - a) / N
    x_k = np.linspace(a, b, N+1)
    integral = (delta_x / 3) * (f(a) + 2*np.sum(f(x_k[2:N:2])) + 4*np.sum(f(x_k[1:N:2]) + f(b)))
    return integral

#Границы и количество разбиений
a = 2
b = 3
N = 10

#Вычисления
integral_rectangle = rectangle_method(f, a, b, N)
integral_trapezoid = trapezoid_method(f, a, b, N)
integral_simpson = simpson_method(f, a, b, N)

integral_rectangle, integral_trapezoid, integral_simpson
```
