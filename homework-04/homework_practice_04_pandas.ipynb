#### 5. Какие 5 товаров были самыми дешёвыми и самыми дорогими? (по item_name)

Для этого будет удобно избавиться от дубликатов и отсортировать товары. Не забудьте про количество товара.

```python
# your code
a=df.drop_duplicates('item_name')
b=(a['item_price_float'])/(a['quantity'])
a=a.assign(item_price_one = b)
a.sort_values(by='item_name',axis=0,ascending=True, inplace=False, kind='quicksort',na_position='last')
max=a.nlargest(5, columns='item_price_one')
min=a.nsmallest(5, columns='item_price_one')

print("Самые дешевые:", min['item_name'], "Самые дорогие", max['item_name'])

```

## 2. Визуализации и matplotlib

При работе с данными часто неудобно делать какие-то выводы, если смотреть на таблицу и числа в частности, поэтому важно уметь визуализировать данные. В этом разделе мы этим и займёмся.

У matplotlib, конечно, же есть [документация](https://matplotlib.org/users/index.html) с большим количеством [примеров](https://matplotlib.org/examples/), но для начала достаточно знать про несколько основных типов графиков:
- plot — обычный поточечный график, которым можно изображать кривые или отдельные точки;
- hist — гистограмма, показывающая распределение некоторое величины;
- scatter — график, показывающий взаимосвязь двух величин;
- bar — столбцовый график, показывающий взаимосвязь количественной величины от категориальной.

В этом задании вы попробуете построить каждый из них. Не менее важно усвоить базовые принципы визуализаций:
- на графиках должны быть подписаны оси;
- у визуализации должно быть название;
- если изображено несколько графиков, то необходима поясняющая легенда;
- все линии на графиках должны быть чётко видны (нет похожих цветов или цветов, сливающихся с фоном);
- если отображена величина, имеющая очевидный диапазон значений (например, проценты могут быть от 0 до 100), то желательно масштабировать ось на весь диапазон значений (исключением является случай, когда вам необходимо показать малое отличие, которое незаметно в таких масштабах).
- сетка на графике помогает оценить значения в точках на глаз, это обычно полезно, поэтому лучше ее отрисовывать.

#### 14. Постройте гистограмму распределения сумм покупок и гистограмму средних цен отдельных видов продуктов item_name.

Изображайте на двух соседних графиках. Для этого может быть полезен subplot.

```python
# your code
import matplotlib.pyplot as plt
import pandas as pd
df =pd.read_csv('chipotle.tsv', sep='\t')
df['item_price'][0]
df['item_price'].apply(lambda s: float(s.strip()[1:]))
df['item_price_float']=df['item_price'].apply(lambda s: float(s.strip()[1:]))
df['item_price_float'].plot(kind='hist', bins=20, color='pink', edgecolor='black')

plt.title('Гистограмма распределения сумм покупок')
plt.xlabel('Стоимость')
plt.ylabel('Частота')
plt.show()

a=df.drop_duplicates('item_name')
srt = a.groupby('item_name')['item_price_float'].mean()

# Построить гистограмму средних цен отдельных видов продуктов
srt.plot(kind='bar', color='pink', edgecolor='black')

plt.title('Гистограмма средних цен отдельных видов продуктов')
plt.xlabel('Имя товара')
plt.ylabel('Средняя цена')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

#### 15. Постройте график зависимости суммы покупок от дней.

```python
# your code
df['date'] = pd.to_datetime(df['date'])
sumdn = df.groupby(df['date'].dt.date)['item_price_float'].sum()
sumdn.plot(color='pink')

plt.title('Зависимость суммы покупок от дней')
plt.xlabel('Дата')
plt.ylabel('Сумма покупок')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### 16. Постройте средних сумм покупок по дням недели (bar plot).

```python
# your code
df['date'] = pd.to_datetime(df['date'])
df['dow'] = df['date'].dt.day_name()
srdow = df.groupby('dow')['item_price_float'].mean()
dow_1 = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
srdow = srdow.reindex(dow_1)
srdow.plot(kind='bar', color='pink', edgecolor='black')

plt.title('Средние суммы покупок по дням недели')
plt.xlabel('День недели')
plt.ylabel('Средняя сумма покупок')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### 17. Постройте график зависимости денег за товар от купленного количества (scatter plot).

```python
# your code
df.plot.scatter(x='quantity', y='item_price_float', color='pink', edgecolor='black')

plt.title('Зависимость денег за товар от купленного количества')
plt.xlabel('Количество товаров')
plt.ylabel('Стоимость')
plt.grid(True)
plt.tight_layout()
plt.show()
```
